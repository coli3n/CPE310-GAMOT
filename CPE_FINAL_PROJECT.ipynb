# 5. **Duration & Seasons.** Parse duration into numeric minutes (for movies) and number of seasons (for shows). What is the average movie length? The average number of seasons per TV show?

```python
netflix['duration'].value_counts()
```

```python
netflix['duration_int'] = netflix['duration'].str.extract('(\d+)').astype(float)
netflix['duration_type'] = netflix['duration'].str.extract('([a-zA-Z]+)')
# Average movie length
avg_movie_length = netflix[netflix['type'] == 'Movie']['duration_int'].mean()
print(f"Average Movie Length: {avg_movie_length:.2f} minutes")
# Average seasons per TV Show
avg_seasons = netflix[netflix['type'] == 'TV Show']['duration_int'].mean()
print(f"Average TV Show Seasons: {avg_seasons:.2f} seasons")
```

In this next set of codes, the first thing that we did is to separate the numerical part and the unit from the 'duration' column into two new columns ('duration_int' and 'duration_type'). Next, we filtered the DataFrame to isolate movies and TV shows to calculate the mean of the numerical duration for movies and the mean of the numerical duration for TV shows (which represents the number of seasons). We then printed out the results so that we can show the average movie length and TV show seasons.

# **6. Genre Analysis.** Which genre has the highest average release year? What might that indicate?

```python
genres = netflix[['title', 'release_year', 'listed_in']].copy()
genres['Genre'] = genres['listed_in'].str.split(', ')
genres = genres.explode('Genre')
avg_year_genre = genres.groupby('Genre')['release_year'].mean().sort_values(ascending=False)
print("Genres with highest average release year:\n")
avg_year_genre.head(10)
```

To determine which genre in the Netflix dataset has the highest average release year, we selected the columns for title, release_year, and listed_in so that we can copy the selected data in it to a new dataframe which we called genres. We split each string in the 'listed_in' column into a list of individual genres and assigned them to a new column named 'Genre' in the genres DataFrame, so for each row in it where the 'Genre' column contains a list, explode() creates new rows, containing the same values from the other columns ('title', 'release_year') but will have one individual genre from the original list in the 'Genre' column. Effectively, it transforms the DataFrame from having multi-genre entries in a list to having one row per title-genre combination.

We then grouped the genres by the unique values in the 'Genre' column to group all rows with the same genre together. After that, we calculated the average of the 'release_year' for all the titles within each genre group and sorted it in descending order to get the highest to lowest mean.


Based on our inference, the data we generated might indicate that genres evolve over the course of time, with well-known genres generating even more better content which could be why the industry kept on releasing them and emerging genres becoming more popular which may lead to a higher average of release year.

# 8. **Rating vs. Type.** Create a cross-tab or pivot table of type vs. rating. Which rating dominates movies compared to TV shows?

```python
netflix['rating'].value_counts()
```

```python
type_rating_ct = pd.crosstab(netflix['type'], netflix['rating'], normalize='index') * 100
print("\nType vs. Rating Cross-tab (%):\n", type_rating_ct.round(2))
```

```python
type_rating_ct.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Comparison of Ratings by Type (%)')
plt.xlabel('Type')
plt.ylabel('Percentage (%)')
plt.legend(title='Rating', bbox_to_anchor=(1.05, 1), loc='upper left') # Move legend outside
plt.xticks(rotation=0) # Keep x-axis labels horizontal
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
```

We first counted the number of occurrences of each unique rating (like TV-MA, PG, R, etc.) in the rating column of the netflix DataFrame.Based on the results, the highest count for the rating is content with TV-MA rating with 3207, followed by TV-14 with 2160 frequency and TV-PG with 863. This helps us understand the distribution of content ratings on Netflix and can show whether Netflix has more mature content (like TV-MA or R) or more family-friendly content (like G or PG), which is useful for analyzing content strategies, answering questions of whether Netflix caters more to adults or children and others.

The cross-tab or contingency table is done to show the percentage distribution of ratings for movies or tv shows. The code normalize='index' means each row (type) adds up to 100%, showing percentage share by type. Using this, we can now compare rating distributions between movies and shows.

Lastly is the stacked bar chart which helps visualize how content ratings are distributed within each type. In the x-axis, we have the labels for the type which are movie and tv show while we have the percentage in the y-axis, which makes the plot readable. The code tight_layout() ensures that labels and the legend don't overlap and the legend is moved outside the plot for clarity.

Overall, this step gives us an idea on the target audiences for each type of content.

# 10. **Aggregations & GroupBy.** For each country, compute the average release year of its titles. Group by year_added and compute the proportion of Movies vs. TV Shows added each year.


```python
# Average release year per country
avg_year_country = netflix.groupby('country')['release_year'].mean().sort_values(ascending=False)
display("Average Release Year by Country:\n", avg_year_country.head(10))

# Proportion of Movies vs. TV Shows per year added
type_by_year = netflix.groupby(['year_added', 'type']).size().unstack(fill_value=0)
type_by_year_ratio = type_by_year.div(type_by_year.sum(axis=1), axis=0)
display("Proportion of Type by Year:\n", type_by_year_ratio.head(10))
```

```python
type_by_year_ratio.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title("Proportion of Movies vs TV Shows Added Each Year")
plt.xlabel("Year Added")
plt.ylabel("Proportion")
plt.legend(title='Type')
plt.tight_layout()
plt.show()
```

For number 10, we wanted to find out which countries are producing newer content on average. First, we grouped the dataset by country and calculated the mean release year for each. Then, we sorted the results in descending order to show which countries have the most recent average release years. This gives us a clear picture of which countries are leading in terms of producing fresh or current content. For example, if South Korea or India is near the top, it could reflect their rapidly growing entertainment industries.

Now we want to explore how the content mix — between Movies and TV Shows — has changed over the years. First, we grouped the data by year_added and type (Movie or TV Show). Then, we counted the number of each type added per year. After that, we normalized the counts to get the proportion of each type, rather than raw counts. We visualized this data using a stacked bar chart, where each bar represents one year. The segments of the bar show what fraction of that year's content additions were Movies vs. TV Shows. This visualization helps us identify trends, such as whether Netflix is shifting its focus toward TV Shows or maintaining a balance over time, pandemic-era effects (e.g., a spike in TV content in 2020 due to halted film production), and even strategic changes in Netflix’s content acquisition.

Through this sequence of filters, groupings, and visualizations, we gain insights into the type of content Netflix is adding, when it's being added, and how content strategies vary across countries and time. # 11. **Applying Functions.** Write a function that, given a director’s name, returns their titles sorted by release_year. Create a reusable function to plot the top N categories for any categorical column.

```python
# Function to return titles by director
def titles_by_director(director_name):
    return netflix[netflix['director'] == director_name][['title', 'release_year']].sort_values(by='release_year')
print("Titles by 'Martin Scorsese':")
display (titles_by_director('Martin Scorsese'))

# Reusable function to plot top N values of any categorical column
def plot_top_n_categories(column_name, n=10):
    top_n = netflix[column_name].value_counts().head(n)
    plt.figure(figsize=(10, 6))
    sns.barplot(x=top_n.values, y=top_n.index, palette='viridis')
    plt.title(f"Top {n} {column_name.capitalize()} Categories")
    plt.xlabel("Count")
    plt.ylabel(column_name.capitalize())
    plt.grid()
    plt.tight_layout()
    plt.show()

# Example usage
plot_top_n_categories('rating', 10)
```

The first function is called titles_by_director. It takes a director's name as input and returns a list of their titles available on Netflix, sorted by release year. Here's what it does: it filters the DataFrame to only include rows where the director matches the name we pass in. Then it selects just the 'title' and 'release_year' columns, and sorts them so we can see the director's work from oldest to newest. For example, if we want to see Martin Scorsese's titles, we call: display(titles_by_director('Martin Scorsese')). This helps us quickly explore a director's contributions to the Netflix library over time.

The second function is a reusable plotting function called plot_top_n_categories. It takes in the name of any categorical column and returns a bar chart of the top N most frequent values. We start by using value_counts() to find the most common values in that column. Then we use Seaborn to draw a horizontal bar chart, showing the count of each category. Here's how we might use it to visualize the most common content ratings: plot_top_n_categories('rating', 10). This gives us a quick visual summary — for example, we might see that 'TV-MA' and 'TV-14' are the most frequent ratings on the platform. So, in summary: the first function helps us look at a director's work chronologically, and the second helps us explore the most common values in any categorical column with just one line of code. These tools make our data exploration more efficient and insightful.
